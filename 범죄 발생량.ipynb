# 전국 5대범죄 발생량 (2011~2020)
https://www.police.go.kr/www/open/publice/publice0202.jsp

```python
import folium
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib as mpl 
import branca.colormap as cm
import matplotlib.pyplot as plt 
%matplotlib inline

plt.rcParams["font.family"] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('5대 범죄 발생 검거 현황.csv')
df
```

```python
df.plot(x="연도", y=["살인", "강도", "강간·강제추행", "절도", "폭력"], kind="line",figsize=(15,8), rot=0, fontsize=15, color = ['#D20000', '#FFA74F', '#FFE12D', '#17A079', '#0F84B5'])
plt.title('전국 5대 범죄 발생 현황 (단위 : 건)', fontsize=17)
plt.semilogy()
plt.legend(loc='upper left',  bbox_to_anchor=(1, 1))
plt.xticks(range(2011,2021,1))
plt.xlabel(None)
plt.show()
plt.savefig('./전국 5대 범죄 발생량.png')
```

```python
df.plot(x="연도", y=["절도", "폭력"], kind="line",figsize=(15,8), rot=0, fontsize=15, color = ['#17A079', '#0F84B5'])
plt.title('전국 5대 범죄 발생 현황 (단위 : 건)', fontsize=17)
plt.legend(loc='upper left')
plt.xticks(range(2011,2021,1))
plt.xlabel(None)
plt.savefig('./전국 5대 범죄(절도 폭력) 발생량.png', transparent=True)
```

```python
df.plot(x="연도", y=["살인", "강도", "강간·강제추행"], kind="line",figsize=(15,8), rot=0, fontsize=15, color = ['#D20000', '#FFA74F', '#FFE12D'])
plt.title('전국 5대 범죄 발생 현황 (단위 : 건)', fontsize=17)
plt.legend(loc='upper left')
plt.xticks(range(2011,2021,1))
plt.xlabel(None)
plt.savefig('./전국 5대 범죄(살인 강도 강간및강제추행) 발생량.png', transparent=True)
```

```python
ax1 = df.plot(x="연도", y=["절도", "폭력"], kind="line", figsize=(15,8), rot=0, fontsize=20, color = ['#17A079', '#0F84B5'])
ax2 = df.plot(x="연도", y=["살인", "강도", "강간·강제추행"], kind="line", figsize=(15,8), rot=0, fontsize=20, color = ['#D20000', '#FFA74F', '#FFE12D'], secondary_y=True, ax=ax1)

# plt.yticks(rotation=45)
# plt.xticks(rotation=45) ## x축 눈금 라벨 설정 - 40도 회전 
plt.title("5대 범죄 발생량 (단위 : 백 건)", size=20)
#ax.set_ylim(0,60)
plt.xticks(range(2011,2021,1))

ax1.set_yticks([150000,200000,250000,300000,350000])
ax1.set_yticklabels(['150', '200', '250', '300', '350'])
ax2.set_yticks([0,5000,10000,15000,20000, 25000])
ax2.set_yticklabels(['0', '5', '10', '15', '20', '25'])


ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('절도, 폭력 발생 건수', size=20)
ax2.set_ylabel('살인, 강도, 강간·강제추행 발생 건수', size=20)
# ax2.axis( ymin = 0, ymax = 30000) 

# ax1.set_yticks()

leg = plt.legend()
ax1.get_legend().remove()
ax2.get_legend().remove()

plt.show()

plt.savefig('./5대 범죄 발생량.png')
```

## 살인

```python
df_살인 = pd.read_csv('서울시 연도별 살인 발생량.csv')
df_살인
```

```python
df_살인['총'] = df_살인.sum(axis=1)
df_살인
```

```python
df_살인.plot(x="연도", y=['총'], kind="line",figsize=(15,8), rot=0, fontsize=15)
plt.title('연도별 살인 발생량 (단위 : 건)', fontsize=17)
plt.legend(loc='upper left')
plt.xticks(np.arange(2015, 2020, 1))
plt.xlabel(None)
plt.show()
```

```python
df_살인.set_index('연도', inplace = True)
del df_살인['총']
df_살인=df_살인.transpose()
df_살인
```

```python
import seaborn as sns
cmap = sns.light_palette('#0B69A3', as_cmap = True)

plt.figure(figsize = (15,15))
sns.heatmap(df_살인, annot = True, cmap = cmap)
plt.title('연도별 살인 발생량')
plt.savefig('./구별 살인 발생량 - 히트맵.png')
```

```python
df_살인 = pd.read_csv('서울시 연도별 살인 발생량_수정.csv')
df_살인
```
