```python
import folium
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib as mpl 
import branca.colormap as cm
import matplotlib.pyplot as plt 
%matplotlib inline
plt.rcParams["font.family"] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('강간 및 강제추행_범죄자와 피해자와의 관계.csv')
df.set_index('연도',inplace = True)
df.fillna(0, inplace=True)  # 공란은 0처리
df
```

```python
del df['계']
df=df.transpose()
df = df.astype(int)
df
```

```python
for i in list(range(2011, 2021)):
    
    fig = plt.figure(figsize=(10, 10)) ## 캔버스 생성
    ax = fig.add_subplot() ## 프레임 생성
    
    df[i].plot(kind='pie', figsize=(12, 12), autopct = '%1.1f%%', fontsize = 13, legend = True) #colors = ['saddlebrown', 'lightpink', 'gold','cornsilk']

    plt.legend(loc='upper right')
    plt.title(f'{i} 년', fontsize = 15)
    plt.show()
```

```python
df['총'] = df.sum(axis=1)
df
```

```python
df['총'].plot(kind='pie', figsize=(12, 12), autopct = '%1.1f%%', fontsize = 13, legend = True)

plt.legend(loc='upper right')
plt.title('범죄자와 피해자와의 관계', fontsize = 15)
plt.show()
```

```python
df = pd.read_csv('강간 및 강제추행_범죄자와 피해자와의 관계 수정 - 복사본.csv')
df.set_index('관계', inplace = True)
df = df.sort_values('2019', ascending=False)
df
```

```python
df['2019'].plot(kind='pie', figsize=(12, 12), autopct = '%1.1f%%', fontsize = 18, legend = True, colors = [ '#035783', '#0B69A3', '#0F84B5', '#1E8FBB',  '#19A1C9',  '#75B8D5',  '#A3CCE0',  '#B8CFD5',  '#E5E9EA'],startangle=160, counterclock=False)

plt.legend(loc='upper right', fontsize=12)
plt.title('범죄자와 피해자와의 관계 (2019년)', fontsize = 20)
plt.savefig('./강간 및 강제추행 범죄자와 피해자와의 관계 (2019년).png')
```
