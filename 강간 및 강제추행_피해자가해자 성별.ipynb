```python
import folium
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import pyplot
import matplotlib.ticker as mticker
%matplotlib inline

import warnings
warnings.filterwarnings(action='ignore')
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
```

## 피해자

```python
df = pd.read_csv('강간 및 강간추행_피해자 성별.csv')
df
```

```python
df.plot(x="연도", y=["남자", "여자"], kind="bar",figsize=(12,8), rot=0, fontsize=15)
plt.title('강간 및 강제추행 피해자 (단위 : 명)', fontsize=17)
plt.xlabel(None)
plt.show()
```

```python
df.set_index('연도', inplace = True)
df=df.transpose()
df['총'] = df.sum(axis=1)
df
```

```python
for i in list(range(2011, 2021)):
    colors = ['blue', 'red']
    
    fig = plt.figure(figsize=(10, 10)) ## 캔버스 생성
    ax = fig.add_subplot() ## 프레임 생성
    
    df[i].plot(kind='pie', figsize=(12, 12), autopct = '%1.1f%%', fontsize = 13, legend = True, colors = colors)

    plt.legend(loc='upper right')
    plt.title(f'{i} 년', fontsize = 15)
    plt.show()
```

```python
df[2019].plot(kind='pie', figsize=(12, 12), autopct = '%1.1f%%', fontsize = 20, legend = True, colors = ['#0B69A3', '#D20000'])
# plt.rc('font', size=18)
# plt.rc('xtick', labelsize=20) # fontsize of the tick labels
# plt.legend(loc='upper right')
plt.title('강간 및 강제추행 피해자 (2019년)', fontsize = 20)
plt.savefig('./강간 및 강제추행 피해자 (2019년).png')
```

```python
df['총'].plot(kind='pie', figsize=(12, 12), autopct = '%1.1f%%', fontsize = 13, legend = True)

plt.legend(loc='upper right')
plt.title('강간 및 강제추행 피해자', fontsize = 15)
plt.show()
```
