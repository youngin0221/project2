# 서울시 범죄발생 장소 (2012~2019)
https://data.seoul.go.kr/dataList/10943/S/2/datasetView.do

```python
import folium
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib as mpl 
import branca.colormap as cm
import matplotlib.pyplot as plt 
%matplotlib inline

plt.rcParams["font.family"] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('강간 및 강제추행_서울시 5대범죄 발생장소.csv')
df=df.transpose() 
df = df.rename(columns=df.iloc[0])  # 1행을 컬럼으로
df = df.drop(df.index[0])  # 1행 삭제
df['총'] = df.sum(axis=1)
df
```

```python
for i in list(range(2012, 2020)):
    
    fig = plt.figure(figsize=(10, 10)) ## 캔버스 생성
    ax = fig.add_subplot() ## 프레임 생성
    
    df[i].plot(kind='pie', figsize=(12, 12), autopct = '%1.1f%%', fontsize = 13, legend = True) #colors = ['saddlebrown', 'lightpink', 'gold','cornsilk']

    plt.legend(loc='upper right')
    plt.title(f'{i} 년', fontsize = 15)
    plt.show()
```

```python
df['총'].plot(kind='pie', figsize=(12, 12), autopct = '%1.1f%%', fontsize = 13, legend = True)

plt.legend(loc='upper right')
plt.title('강간 및 강제추행 발생 장소 (서울시)', fontsize = 15)
plt.show()    # 노상 > 교통수단 > 유흥접객업소 > 숙박업소, 목욕탕 > 단독주택 > 아파트, 연립다세대 > 역대합실 > 사무실 > 상점 > 유원지
```

```python
df = pd.read_csv('강간 및 강제추행_서울시 5대범죄 발생장소 - 복사본.csv')
df=df.transpose() 
df = df.rename(columns=df.iloc[0])  # 1행을 컬럼으로
df = df.drop(df.index[0])  # 1행 삭제
df = df.sort_values(2019, ascending=False)  # 2019년 기준으로 정렬
df
```

```python
df[2019].plot(kind='pie', figsize=(12, 12), autopct = '%1.1f%%', fontsize = 13, legend = True, colors = [ '#035783', '#0B69A3', '#0F84B5', '#1E8FBB',  '#19A1C9',  '#75B8D5'],startangle=90, counterclock=False)

plt.legend(loc='upper right')
plt.title('범죄발생 장소 (2019년)', fontsize = 15)
plt.savefig('./강간 및 강제추행 서울시 범죄발생 장소 (2019년).png')
```
