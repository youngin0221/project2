# 인구 천 명당 강간및강제추행 발생건수
https://kosis.kr/statHtml/statHtml.do?orgId=135&tblId=DT_13501N_A120 - 범죄발생 지역

https://kosis.kr/statHtml/statHtml.do?orgId=101&tblId=DT_1B040A3&checkFlag=N - 행정구역(시군구)별, 성별 인구수

```python
import folium
import json

geo_str = json.load(open('TL_SCCO_CTPRVN.json', encoding='utf8'))
```

```python
import pandas as pd 
df = pd.read_csv('인구 천명당 강간및강제추행 발생건수.csv')
df
```

```python
map = folium.Map(
    location=[36.45, 127.42], 
    zoom_start=7, tiles='cartodb positron'
)
folium.Choropleth(
    geo_data=geo_str,
    data=df,
    columns=('시도별','인구 천명당 범죄발생건수'),
    fill_color='Blues',
    key_on='feature.properties.CTP_KOR_NM',
    fill_opacity=1,
    line_opacity=0.6,
    legend_name='인구 천 명당 강간 및 강제추행 발생건수 (단위 : 건)'
).add_to(map)
map.save('map.html')
map
```
