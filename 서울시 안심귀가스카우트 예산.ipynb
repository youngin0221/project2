```python
import pandas as pd 
df = pd.read_csv('서울시 안심귀가스카우트 예산.csv')
df
```

```python
df.plot(x="년도", y="예산 (억원)", kind="bar",figsize=(12,8), title = "년도별 서울시 안심귀가스카우트 예산")
```

```python
year = df['년도'].values.tolist()
money = df['예산 (억원)'].values.tolist()
```

```python
plt.rcParams['figure.figsize'] = (15,8)
plt.figure()

plt.bar(year, money)
# plt.legend()

# plt.xlabel("년도", size=17)
plt.xticks(size=17) ## x축 눈금 라벨 설정 - 40도 회전 
# plt.ylabel("예산 (억원)", size=17)
# plt.ylim(3000,6000)
# plt.yticks([3000, 4000, 5000, 6000],size=17)
plt.yticks(size=17)

plt.title("연도별 서울시 안심귀가스카우트 예산 (단위 : 억원)", size=17)

for i, v in enumerate(year):
    plt.text(v, money[i], money[i],                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
             fontsize = 13, 
             color='#ff6600',
             horizontalalignment='center',  # horizontalalignment (left, center, right)
             verticalalignment='bottom')    # verticalalignment (top, center, bottom)

str_val = '%d억원' % money[i]
plt.savefig('./서울시 안심귀가스카우트 예산.png')
```
